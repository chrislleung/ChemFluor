# Batch Prediction

Use this notebook to predict multiple candidate molecules from a CSV file.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

sys.path.insert(0, str(PROJECT_ROOT))
print("Project root:", PROJECT_ROOT)
```

## Create an Example Candidate CSV

The CSV must contain `SMILES` and `solvent`. A `name` column is optional.

```python
import pandas as pd

candidates = pd.DataFrame([
    {"name": "benzene_example", "SMILES": "c1ccccc1", "solvent": "MeCN"},
    {"name": "ethanol_example", "SMILES": "CCO", "solvent": "MeOH"},
    {"name": "triethylamine_example", "SMILES": "CCN(CC)CC", "solvent": "MeCN"},
])

candidate_path = PROJECT_ROOT / "example_candidates.csv"
candidates.to_csv(candidate_path, index=False)
print("Saved:", candidate_path)
candidates
```
