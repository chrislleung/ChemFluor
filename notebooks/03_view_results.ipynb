# View Training Results

This notebook loads saved metrics, plots, and worst-prediction tables from the `outputs/` folder.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

sys.path.insert(0, str(PROJECT_ROOT))
print("Project root:", PROJECT_ROOT)
```

```python
import pandas as pd
from pathlib import Path
from IPython.display import Image, display

outputs = PROJECT_ROOT / "outputs"
metrics_dir = outputs / "metrics"
plots_dir = outputs / "plots"
models_dir = outputs / "models"

print("Metrics dir:", metrics_dir)
print("Plots dir:", plots_dir)
print("Models dir:", models_dir)
```

## Metrics Table

```python
metrics_path = metrics_dir / "metrics.csv"
if metrics_path.exists():
    metrics = pd.read_csv(metrics_path)
    display(metrics)
else:
    print("metrics.csv not found. Run training first.")
```

## Key Plots

```python
for filename in [
    "predicted_vs_actual_wavelength.png",
    "residuals_vs_actual_wavelength.png",
    "predicted_vs_actual_plqy.png",
    "residuals_vs_actual_plqy.png",
    "confusion_matrix_plqy_classifier.png",
]:
    path = plots_dir / filename
    if path.exists():
        print(filename)
        display(Image(str(path)))
    else:
        print("Missing:", filename)
```

## Worst Wavelength Predictions

```python
path = plots_dir / "worst_20_wavelength_predictions.csv"
if path.exists():
    display(pd.read_csv(path))
else:
    print("Worst wavelength prediction file not found.")
```

## Worst PLQY Predictions

```python
path = plots_dir / "worst_20_plqy_predictions.csv"
if path.exists():
    display(pd.read_csv(path))
else:
    print("Worst PLQY prediction file not found.")
```

## Uncertainty Files

```python
for filename in ["wavelength_uncertainty.csv", "plqy_uncertainty.csv"]:
    path = metrics_dir / filename
    if path.exists():
        print(filename)
        display(pd.read_csv(path).head())
    else:
        print("Missing:", filename)
```
