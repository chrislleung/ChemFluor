# ChemFluor Project Overview

This notebook is a non-technical overview of the ChemFluor molecular property prediction pipeline.

The project predicts fluorescent molecule properties from:

- `SMILES`: text representation of molecular structure
- `solvent`: solvent used for the fluorescence measurement

The model predicts:

- emission wavelength in nm
- PLQY as a continuous brightness value
- bright/dim class using `PLQY > 0.25`

## Why This Project Exists

The goal is to recreate and extend a ChemFluor-style machine learning workflow so that chemists can screen candidate fluorescent molecules before experimental testing.

The model is not meant to replace experiments. It is meant to help prioritize candidates and flag when predictions are uncertain.

## Model Inputs

The model uses two categories of inputs.

### Molecular features

- Morgan fingerprints
- MACCS keys
- RDKit molecular descriptors

These describe local molecular fragments, rings, polarity, size, hydrogen bonding, aromaticity, and molecular complexity.

### Solvent features

- one-hot solvent identity
- physical solvent descriptors, when available

The solvent descriptor columns are:

```text
dielectric_constant
refractive_index
dipole_moment
hbond_donor
hbond_acceptor
polarity_ET30
```

## Validation Strategy

The pipeline reports two kinds of model performance.

### Random split

Random split measures interpolation. It answers: "How well does the model perform on molecules similar to molecules it has already seen?"

### Scaffold split

Scaffold split is stricter. It separates Bemis-Murcko scaffolds between training and testing. It answers: "How well does the model perform on new molecular families?"

For general lab use, scaffold split is the more important score.

## Applicability Domain

When predicting a new molecule, the model checks whether that molecule is similar to the training data.

The report includes:

- max Tanimoto similarity to training molecules
- top-5 nearest-neighbor similarity
- whether the scaffold was seen during training
- whether the solvent was seen during training
- confidence level: High, Medium, or Low

Low-confidence predictions should be treated as rough screening estimates.
